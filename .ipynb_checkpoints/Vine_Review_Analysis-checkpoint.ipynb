```python
import pandas as pd
```

```python
#Read CSV file
vine_table = pd.read_csv("vine_table.csv")
vine_table
```

```python
total_votes = vine_table.loc[vine_table["total_votes"] >= 20]
total_votes
```

```python
total_votes["helpful_percent"] = total_votes["helpful_votes"]/total_votes["total_votes"] 
total_votes = total_votes.reset_index()
total_votes
```

```python
type(total_votes["helpful_percent"][0])
```

```python
total_votes_df = total_votes.loc[total_votes["helpful_percent"] >= .50]
total_votes_df
```

```python
paid_program = total_votes_df.loc[total_votes_df["vine"] == 'Y']
paid_program
```

```python
unpaid_program = total_votes_df.loc[total_votes_df["vine"] == 'N']
unpaid_program
```

```python
paid_program_count = paid_program['vine'].value_counts()
paid_program_count
```

```python
paid_program_5_star = paid_program.loc[paid_program["star_rating"] == 5]
paid_program_5_star.count()
```

```python
unpaid_program_count = unpaid_program['vine'].value_counts()
unpaid_program_count
```

```python
unpaid_program_5_star = unpaid_program.loc[unpaid_program["star_rating"] == 5]
unpaid_program_5_star.count()
```

```python
paid_5_percent = (paid_program_5_star["star_rating"].count()/paid_program_count)*100
paid_5_percent
```

```python
unpaid_5_percent = (unpaid_program_5_star["star_rating"].count()/unpaid_program_count)*100
unpaid_5_percent
```
